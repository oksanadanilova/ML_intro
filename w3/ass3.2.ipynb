```python
#Downloading DS
newsgroups = datasets.fetch_20newsgroups(subset='all', categories= ['alt.atheism', 'sci.space'])
y = newsgroups.target
```

```python
#TF_IDF Vectorizing
vector = TfidfVectorizer()
x = vector.fit_transform(newsgroups.data)
idf = vector.idf_
```
