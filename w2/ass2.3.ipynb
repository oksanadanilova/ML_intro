```python
import pandas
```

```python
from sklearn.linear_model import Perceptron
```

```python
import numpy as np
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
f = open('perceptron-train.csv')
data = pandas.read_csv(f, names=['y', 'x1', 'x2'])
f.close()
```

```python
data.head()
```

```python
y_train = data['y']
y_train
```

```python
y_train = y_train.as_matrix()
```

```python
y_train
```

```python
x_train = data.drop(['y'], axis=1)
x_train
```

```python
x_train = x_train.as_matrix()
x_train
```

```python
f = open('perceptron-test.csv')
data_test = pandas.read_csv(f, names=['y', 'x1', 'x2'])
f.close()
y_test = data_test['y']
y_test = y_test.as_matrix()
x_test = data_test.drop(['y'], axis=1)
x_test = x_test.as_matrix()
```

```python
p = Perceptron(random_state=241)
```

```python
p.fit(x_train, y_train)
```

```python
predictions = p.predict(x_test)
```

```python
score_ass1 = accuracy_score(y_test, predictions)
```

```python
score_ass1
```

```python
x_train_norm = scaler.fit_transform(x_train)
x_test_norm = scaler.transform(x_test)
```

```python
p_norm = Perceptron(random_state=241)
p.fit(x_train_norm, y_train)
predictions_norm = p.predict(x_test_norm)
score_ass2 = accuracy_score(y_test, predictions_norm)
```

```python
score_ass = score_ass2 - score_ass1
score_ass
```

```python
f = open('ans6', 'w')
f.write(str(round(score_ass, 3)))
f.close()
```
